```
X = np.load('Data.npy')
Y = np.load('Labels.npy')
```

```
import numpy as np
for k in range(7):
    sume = np.zeros(90)
    count = 0 
    for i in range(X.shape[0]):
        if(Y[i][k] == 1):
            print(i)
            count = count + 1
            for j in range(500):
                  sume = sume + X[i][j]
    print("count ",count,sume)
    count = count * 500
    avg = sume/count
    print("avg ",avg)
                    
    for i in range(X.shape[0]):
        if(Y[i][k] == 1):
            for j in range(500):
                X[i][j] = X[i][j] - avg
```

```
import numpy as np
import matplotlib.pyplot as plt
from pandas import read_csv
from keras.models import Sequential
from keras.layers import Dense,Dropout,BatchNormalization
from keras.layers import LSTM
from sklearn.preprocessing import MinMaxScaler
from sklearn.metrics import mean_squared_error
from sklearn.model_selection import train_test_split
from keras.callbacks import ModelCheckpoint
from plot_confusion import plot_confusion_matrix
#from keras.utils import plot_model

#parameters
n_hidden = 200
time_steps = 500
features = 90
n_epochs = 80
batch_size = 200
learning_rate = 0.0001
display_step = 100

#model Architecture
model = Sequential()
model.add(LSTM(n_hidden, input_shape=(time_steps,features)))
model.add(BatchNormalization())
#model.add(Dropout(0.2))
model.add(Dense(7,activation='softmax'))     #7 classes not including no_activity
model.summary()
#plot_model(model, to_file='model.png')


#X = np.load('Data.npy')
#Y = np.load('Labels.npy')
#X = np.load('Data_phase.npy')
#Y = np.load('Labels_phase.npy')
#p = np.load('phase_off.npy')
#for i in range(X.shape[0]):
#    for j in range(500):
#        X[i][j] = X[i][j] - p
print(X.shape,Y.shape)
X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size=0.20, random_state=0)
print(X_train.shape,y_train.shape)
model.compile(loss='categorical_crossentropy', optimizer='adam', metrics=['accuracy'])
filepath="sub_avg_amp_y-{epoch:02d}-{val_acc:.2f}.hdf5"
checkpoint = ModelCheckpoint(filepath, monitor='val_acc', verbose=1, save_best_only=True, mode='max')
callbacks_list = [checkpoint]
history = model.fit(X_train,y_train,validation_data=(X_test,y_test), epochs=n_epochs,callbacks=callbacks_list,batch_size=batch_size,verbose=2,shuffle=True)

print(history.history.keys())
# summarize history for accuracy
np.save('avg_train_accu_amp.npy',history.history['acc'])
np.save('avg_val_accu_amp.npy',history.history['val_acc'])
plt.plot(history.history['acc'])
plt.plot(history.history['val_acc'])
plt.title('model accuracy')
plt.ylabel('accuracy')
plt.xlabel('epoch')
plt.legend(['train', 'vali'], loc='upper left')
#plt.savefig('accu_amplitude.eps')
plt.show()
# summarize history for loss
plt.figure()
plt.plot(history.history['loss'])
plt.plot(history.history['val_loss'])
plt.title('model loss')
plt.ylabel('loss')
plt.xlabel('epoch')
plt.legend(['train', 'vali'], loc='upper left')
#plt.savefig('loss_phasereal_denoised.eps')
plt.show()
np.save('avg_train_loss_amp.npy',history.history['loss'])
np.save('avg_val_lossd_amp.npy',history.history['val_loss'])
```
